### Mean Squred Error: a cost function for regression problem 

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm #Color_Map
```

```python
# make sample data
x=np.array([[1, 1.6, 1.3, 2, 5, 4,5.4, 9.2,3]]).transpose()
y=np.array([2,5,3,6,5.6,1,8,3,9]).reshape(9,1)
print('x:',x.shape)
print('y:',y.shape)
```

```python
r=LinearRegression()
r.fit(x,y)
print('theta_zero:',r.intercept_[0])
print('theta_one:',r.coef_[0][0])
```

```python
plt.scatter(x,y,s=50)
plt.plot(x,r.predict(x))
```
