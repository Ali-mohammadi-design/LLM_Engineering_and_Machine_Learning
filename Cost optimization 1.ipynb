```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Example 1
f(x)=x^2 +x +1

```python
def f(x):
    return x**2 + x+1
```

```python
f(1)
```

```python
#data generating
x_1=np.linspace(start=-300, stop=300, num=10000)
```

```python
x_1
```

```python
x_1[0]
```

```python
g2=1
for i in range (0,10000):
    g1=abs(f(x_1[i]))
    if g2>=g1:
        g2=g1
        print(f'x_1[{i}]={x_1[i]} and g1 = {g1}')
print(f'the minmum f(x) is {g2}')
```

```python
#plot
plt.plot(x_1,f(x_1))
```

```python
#plot
plt.xlim([-3,3])
plt.ylim([-.2,2])
plt.xlabel('X',fontsize=17)
plt.ylabel('f(x)',fontsize=17)
plt.plot(x_1,f(x_1))
```
