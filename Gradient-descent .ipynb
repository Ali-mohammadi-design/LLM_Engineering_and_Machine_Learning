## gradient descent as a python function

```python
def grad(dg_function, start,step=0.2,prec=0.1):

    new=start
    previous=0.0
    for i in range(1000000):
        previous=new
        d1=dg_function(previous)
        new=previous-step*d1
        difference= abs(new-previous)
        if difference<=prec:
         break
        return new
```

```python
new=grad(dg,1,0.1,0.001)
print("x",new)
print("g(x)=",g(new))
```
