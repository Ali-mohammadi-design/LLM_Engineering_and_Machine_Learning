```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
f=pandas.read_csv("cost_revenue_clean.csv")
```

```python
f
```

```python
f.describe()
```

```python
x=f["production_budget_usd"]
y=f["worldwide_gross_usd"]
```

```python
x2= DataFrame(f,columns=["production_budget_usd"])
```

```python
y2=DataFrame(f,columns=["worldwide_gross_usd"])
```

```python
y2
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x,y,alpha=0.3)
plt.title("production cost and revenue for films")
plt.xlabel("production budget in $")
plt.ylabel("revenues in $")
plt.ylim(0,)
plt.xlim(0,)
```

```python
r=LinearRegression()
r.fit(x2,y2)
```

slope coef:

```python
r.coef_ #tetha_1
```

```python
r.intercept_
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x,y,alpha=0.3)
plt.plot(x2,r.predict(x2), color='red', linewidth=4)
plt.title("production cost and revenue for films")
plt.xlabel("production budget in $")
plt.ylabel("revenues in $")
plt.ylim(0,)
plt.xlim(0,)
```

```python
r.score(x2,y2)
```
